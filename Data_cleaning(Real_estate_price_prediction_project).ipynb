```python
def is_float(x):
    try:
       float(x)
    except:
      return False
    return True
```

```python
df[~df['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens) == 2:
    return (float(tokens[0])+float(tokens[1]))/2
  try:
      return float(x)
  except:
      return None
```

```python
convert_sqft_to_num('21660')
```

```python
convert_sqft_to_num('1145 - 1340')
```

```python
convert_sqft_to_num(type('34.46Sq. Meter'))
```

```python
df = df.copy()
df['total_sqft'] = df['total_sqft'].apply(type('convert_sqft_to_num'))
df.head(3)
```
